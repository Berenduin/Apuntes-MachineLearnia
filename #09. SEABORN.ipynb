Seaborn es una biblioteca escrita sobre la base de Matplotlib y Pandas, y que nos permite la visualización de datos de forma rápida y muy práctica.

Vídeo: https://www.youtube.com/watch?v=xYgfIRzNPlo

Documentación: https://seaborn.pydata.org/index.html

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('iris.csv')
```

```python
iris.head()
```

```python
plt.scatter(iris['sepal.length'], iris['sepal.width'])
plt.show()
```

```python
# veamos las capacidades de seaborn
import seaborn as sns
sns.pairplot(iris)
```

```python
# podemos añadir formato y color a nuestros gráficos
sns.pairplot(iris, hue='variety')
```

Ejemplos de otras representaciones: https://seaborn.pydata.org/examples/index.html

API: https://seaborn.pydata.org/api.html

En resumidas cuentas la estructura en el módulo de seaborn sería: 

    sns.nombre_funcion(x, y, datos, hue, size, style)

```python
titanic =  sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.columns
```

```python
titanic.drop(['class', 'who', 'adult_male', 'embark_town','alive', 'alone'], axis=1, inplace=True)
titanic.dropna(axis=0, inplace=True)
titanic.head()
```

```python
sns.pairplot(titanic)
```

Para este caso del Titanic, el análisis que nos ofrece sns.pairplot() no genera tablas que nos devuelvan mucha información relevante. Esto se debe a que muchos campos contienen datos cualitativos y por lo tanto nos haría falta realizar un análisis de categorías más que un análisis cuantitativo como el método que estabamos usando.

```python
sns.catplot(x='pclass', y='age', data=titanic, hue='sex')
plt.show()
```

```python
sns.boxplot(x='pclass', y='age', data=titanic, hue='sex')
plt.show()
```

```python
sns.distplot(titanic['fare'])
plt.show()
```

```python
sns.jointplot('age', 'fare', data=titanic, kind = 'kde')
plt.show()
```

```python
sns.jointplot('age', 'fare', data=titanic, kind = 'hex')
plt.show()
```

```python
sns.heatmap(titanic.corr())
```
